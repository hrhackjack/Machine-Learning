```python
data = {'State':['CA','NY','LA','CA','LA','CA'],
       'Person':['Sam','Steve','Stark','Carol','Thor','Natasha'],
       'Sales':[200,150,375,135,265,450]}
```

```python
dfd = pd.DataFrame(data)
```

```python
per = dfd.groupby('Person')
```

```python
per.mean()
```

```python
per.sum()
```

```python
per.sum().loc['Stark']
```

```python
dfd.groupby("State").sum().loc["LA"]
```

```python
dfd.groupby("State").count()
```
